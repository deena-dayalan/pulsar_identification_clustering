```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.svm import SVC
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold, StratifiedKFold, RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
## dataset file
pulsar_dataset = '/home/d.dasarathan/ds5500/projects/datasets/HTRU2/HTRU_2.csv'
```

```python
## read dataset
df_main = pd.read_csv(pulsar_dataset, header=None)
```

```python
## display sample dataset
df_main.head()
```

```python
## add column names to dataset
col_names = ['mean_int_pf', 'std_pf', 'ex_kurt_pf', 'skew_pf', 'mean_dm', 'std_dm', 'kurt_dm', 'skew_dm','label']
df_main.columns = col_names
df_main.head()
```

```python
df_main.shape
```

```python
df_main.label.value_counts()
```

```python
df_main['label'].value_counts()/np.float(len(df_main)) * 100
```

```python
#X = df_main.drop(['label'], axis=1)
X = df_main[['mean_int_pf', 'std_pf', 'ex_kurt_pf', 'skew_pf', 'mean_dm', 'std_dm', 'kurt_dm', 'skew_dm']]
y = df_main['label']
```

```python
X.head()
```

```python
y.head()
```

```python
## split X and y into train and test sets
## 75%-25% split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
print("X train shape : ",X_train.shape)
print("y train shape : ",y_train.shape)
print("X test shape : ",X_test.shape)
print("y test shape : ",y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
## train dataset label distribution
y_train.value_counts()
```

```python
## test dataset label distribution
y_test.value_counts()
```

----------------------------------------------------------------------------------------------------

#### Scaling Features

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
#X_train = pd.DataFrame(X_train, columns=[col_names])
```

```python
#X_test = pd.DataFrame(X_test, columns=[col_names])
```

------------------------------------------------------------------------------------

```python
## Array to maintain the model results
model_results = []
```

#### SVM (default hyperparameters; kernel=rbf and C=1.0)

```python
rbf_svc=SVC()

rbf_svc.fit(X_train,y_train)

y_pred_svm_train = rbf_svc.predict(X_train)

y_pred_svm_test = rbf_svc.predict(X_test)

print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_svm_train)))

print('Model accuracy score : {0:0.4f}'. format(accuracy_score(y_test, y_pred_svm_test)))

## Confusion Matrix
cm = confusion_matrix(y_test, y_pred_svm_test)

print('Confusion matrix\n', cm)
print('True Positives(TP) = ', cm[1,1])
print('True Negatives(TN) = ', cm[0,0])
print('False Positives(FP) = ', cm[0,1])
print('False Negatives(FN) = ', cm[1,0])

TP = cm[1,1]
TN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]

## classification accuracy
classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)
print('Classification accuracy : {0:0.4f}'.format(classification_accuracy))

## classification error
classification_error = (FP + FN) / float(TP + TN + FP + FN)
print('Classification error : {0:0.4f}'.format(classification_error))

## precision score
precision = TP / float(TP + FP)
print('Precision : {0:0.4f}'.format(precision))

## recall 
recall = TP / float(TP + FN)
print('Recall or Sensitivity : {0:0.4f}'.format(recall))

## specificity
specificity = TN / (TN + FP)
print('Specificity : {0:0.4f}'.format(specificity))

## F-1 score
f1 = 2 * (precision * recall) / (precision + recall)
print('F-1 score : {0:0.4f}'.format(f1))

ROC_AUC = roc_auc_score(y_test, y_pred_svm_test)
print('ROC AUC : {:.4f}'.format(ROC_AUC))

## Store model results in Dictionary
svm_1 = dict()
svm_1['Model'] =  'SVC'
svm_1['Kernel'] =  'rbf'
svm_1['C'] =  1.0
svm_1['Classification_Accuracy'] =  classification_accuracy
svm_1['Classification_Error'] =  classification_error
svm_1['Precision'] =  precision
svm_1['Recall'] =  recall
svm_1['F1'] =  f1
svm_1['ROC_AUC'] =  ROC_AUC
model_results.append(svm_1)
```

```python
## Confusion matrix heatmap

cm_matrix = pd.DataFrame(data=cm, columns=['Predict: RFI', 'Predict: Pulsar'], 
                                 index=['Actual:RFI', 'Actual:Pulsar'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

#### Hyperparameter tuning using GridSearchCV
